```python
import pandas as pd
import networkx as nx
import numpy as np
from scipy.stats import spearmanr, kendalltau

# Load the dataset
file_paths = [
    "Mutated_Genes (1).txt",
    "Mutated_Genes (2).txt",
    "Mutated_Genes.txt"
]

# Read and merge datasets
dfs = [pd.read_csv(file, sep="\t", engine="python") for file in file_paths]
merged_df = pd.concat(dfs, ignore_index=True)

# Convert mutation frequency to numerical values
merged_df["Freq"] = merged_df["Freq"].str.rstrip("%").astype(float)

# Create a directed graph
G = nx.DiGraph()
for _, row in merged_df.iterrows():
    gene = row["Gene"]
    freq = row["Freq"]
    G.add_node(gene, weight=freq)

# Define ground truth sorting based on mutation frequency
ground_truth = sorted(G.nodes, key=lambda gene: G.nodes[gene]["weight"], reverse=True)[:20]  # Top 20 genes

# Simulated QAOA sorting approach (assumed to be sorted by frequency)
sorted_genes = sorted(G.nodes, key=lambda gene: G.nodes[gene]["weight"], reverse=True)

# Extract the top 20 predicted genes from QAOA sorting
predicted_genes = sorted_genes[:20]

# Convert gene names to index positions for evaluation
true_labels = np.arange(len(ground_truth))  # Assign true ranking positions
predicted_labels = [ground_truth.index(gene) if gene in ground_truth else -1 for gene in predicted_genes]

# Remove invalid entries (genes not in ground truth)
valid_indices = [i for i in range(len(predicted_labels)) if predicted_labels[i] != -1]
true_labels = np.array(true_labels)[valid_indices]
predicted_labels = np.array(predicted_labels)[valid_indices]

# Compute Spearman's rank correlation
spearman_corr, spearman_p_value = spearmanr(true_labels, predicted_labels)

# Compute Kendall's Tau correlation
kendall_tau, kendall_p_value = kendalltau(true_labels, predicted_labels)

# Display results
print("Quantum Sorting Model Evaluation Scores:")
print(f"Spearman's Rank Correlation: {spearman_corr:.4f} (p-value: {spearman_p_value:.4e})")
print(f"Kendall's Tau: {kendall_tau:.4f} (p-value: {kendall_p_value:.4e})")
```

```python
# Import necessary libraries
import numpy as np
import pandas as pd
import time
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from qiskit.utils import algorithm_globals
from qiskit.circuit.library import ZZFeatureMap
from qiskit_machine_learning.kernels import QuantumKernel
from qiskit_aer import Aer
from qiskit_machine_learning.algorithms import QSVC

# Set random seed for reproducibility
algorithm_globals.random_seed = 42

# ✅ **Prepare Quantum Kernel and Feature Map**
feature_map = ZZFeatureMap(feature_dimension=2, reps=2)
quantum_kernel = QuantumKernel(feature_map=feature_map, quantum_instance=Aer.get_backend("qasm_simulator"))

# ✅ **Convert Mutation Data into Quantum Features**
freq_values = np.array([G.nodes[gene]["weight"] for gene in top_20_genes])

# Ensure feature_dimension >= 2 by adding log-transformed mutation frequency
normalized_freq = (freq_values - np.min(freq_values)) / (np.max(freq_values) - np.min(freq_values))  # Normalize
log_freq = np.log1p(freq_values)  # Log transform
top_20_gene_frequencies = np.column_stack((normalized_freq, log_freq))  # Combine both features

# Generate labels
y_labels = np.arange(len(top_20_genes))  # Assign labels for sorting

# ✅ **Measure time complexity: Start time**
start_time = time.time()

# ✅ **Train QSVC Quantum Classifier**
qsvc = QSVC(quantum_kernel=quantum_kernel)
qsvc.fit(top_20_gene_frequencies, y_labels)

# ✅ **Predict labels using quantum model**
predicted_labels = qsvc.predict(top_20_gene_frequencies)

# ✅ **Measure time complexity: End time**
end_time = time.time()
execution_time = end_time - start_time

# ✅ **Compute accuracy metrics**
accuracy = accuracy_score(y_labels, predicted_labels)
precision = precision_score(y_labels, predicted_labels, average="weighted", zero_division=0)
recall = recall_score(y_labels, predicted_labels, average="weighted", zero_division=0)
f1 = f1_score(y_labels, predicted_labels, average="weighted", zero_division=0)

# ✅ **Display accuracy measures and time complexity**
accuracy_measures = {
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1,
    "Execution Time (seconds)": execution_time
}

# Print accuracy metrics and execution time
print("\nQuantum Model Accuracy and Time Complexity:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1:.4f}")
print(f"Execution Time: {execution_time:.4f} seconds")
```
